```python
import random  
from random import choice
import pandas as pd
import numpy as np 

#Function that makes a random sequence
def random_dna_sequence(length):  

  

    return ''.join(random.choice('ACTG') for _ in range(length))  

  
#Makes 10,000 sequences
for _ in range(1):  

    dna = random_dna_sequence(500000) 

n=50 #sequence length to divide the dna size by

#separates 500000nt into list of 10000 50nt sequences
sequences_10000 = [dna[i:i+n] for i in range(0, len(dna), n)] 

#Makes empty list and string of As
string = 'AAAAAAAAAA'
random_sequences_1000=[]

#adds dna value to list 1 time
while len(random_sequences_1000) != 1000:
    for _ in range(1):  
        dna = random_dna_sequence(50)
        string= 'AAAAAAAAAA'
        #create index range for replacement
        #after position 40, some sequences will end up with 50+ bps
        max_random = random.randint(0,40)

       #replaces portion of sequences with 10As at random spot
        dna= dna.replace(dna[max_random:max_random+10], string)
        random_sequences_1000.append(dna)

random_sequences = random_sequences_1000+ sequences_10000

#shuffle sequences and put in a dataframe
random.shuffle(random_sequences)

random_sequences=pd.DataFrame(random_sequences) 

#adding total_reads column
total_reads = []
for _ in range (0,11000):
   i= random.randint(0,12000)
   total_reads.append(i)
random_sequences['total_reads']=total_reads

#adding rl column
rl = []
for _ in range (0,11000):
   i= np.random.normal(5.000000000, 1.000000000)
   rl.append(i)
random_sequences['rl'] = rl


random_sequneces = random_sequences.rename(columns= {0: 'utr'} , inplace = True)

#make polyA sequences have a scaled_rl of 5
random_sequences['rl'] = np.where (random_sequences.utr.str.contains("AAAAAAAAAA", regex=False), 5,random_sequences.rl)

random_sequences.to_csv('random_sequences.csv')
                        
random_sequences.to_pickle('random_sequences.pkl')

print(random_sequences.tail(10))
```
