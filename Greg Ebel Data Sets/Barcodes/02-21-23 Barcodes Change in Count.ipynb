```python
#load data sets 

df=pd.read_csv("CSV Files/bcWNV_P2_50_S309_barcode_counts.csv", index_col=0) 

  

df2=pd.read_csv("CSV Files/cxt_mg_12dpi_1_S62_barcode_counts.csv", index_col=0) 

  

df3=pd.DataFrame() 

  

#Make several empty lists to fill with dictionary values 

list1=[] #the sequences 

list2=[] #the input counts 

list3=[] #the 12dpi counts 

list4=[] #the difference in the counts 

Input= df.set_index('Barcode Sequence')['Count'].to_dict() 

MG= df2.set_index('Barcode Sequence')['Count'].to_dict() 

  

#fill the lists 

for x,y in Input.items(): 

    if x in MG.keys(): 

        list1.append(x) 

        list2.append(y) 

        a=MG.get(x) 

        list3.append(a) 

  

#make list 3 the difference between the counts 

for a,b in zip(list3,list2): 

    list4.append(a-b) 

  

  

#make a dictionary with the duplicate sequences and the differences between the barcodes 

New_Counts={} 

  

for key in list1: 

    for value in list4: 

        New_Counts[key] = value 

        list4.remove(value) 

        break 

         

  

#Now lets turn the dictioanry into a dataframe 

  

df3["Barcode Sequence"]=New_Counts.keys() 

  

df3["Count"] = New_Counts.values() 

  

#Make all values that are not in midgut negative to apply for lack of count 

df['Count'] = df['Count'].apply(lambda x: x*-1) 

  

  

# merge df3 (which has count differences) with input df and remove duplicates 

data=(df3,df) 

df4=pd.concat(data, ignore_index=True) 

  

df5=df4.drop_duplicates(subset = "Barcode Sequence", keep="first") 

  

#save files 

df5.to_csv("Alligned Barcode/Barcode_Input_vs_mg.csv") 

df5.to_pickle("Alligned Barcode/Barcode_Input_vs_mg.pkl") 

print (df5) 

 
```
