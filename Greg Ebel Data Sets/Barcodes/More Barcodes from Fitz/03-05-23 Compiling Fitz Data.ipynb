```python
import pandas as pd
import numpy as np

#read the csv files, delimit them by tab, and make it so that the header is not the first value point in the data
df=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_1_S62_true_barcodes.txt", sep='\t', header = None)

df2=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_10_S71_true_barcodes.txt", sep='\t', header = None)

df3=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_2_S63_true_barcodes.txt", sep='\t', header = None)

df4=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_3_S64_true_barcodes.txt", sep='\t', header = None)

df5=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_4_S65_true_barcodes.txt", sep='\t', header = None)

df6=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_5_S66_true_barcodes.txt", sep='\t', header = None)

df7=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_6_S67_true_barcodes.txt", sep='\t', header = None)

df8=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_7_S68_true_barcodes.txt", sep='\t', header = None)

df9=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_8_S69_true_barcodes.txt", sep='\t', header = None)

df10=pd.read_csv("More Barcodes From Fitz/Original Data Sets/cxt_mg_12dpi_9_S70_true_barcodes.txt", sep='\t', header = None)


 
#rename the columns so that they have meaning
#create a column for frequency
df=pd.DataFrame(df)
df.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df["Frequency"]=df["Count"]/np.sum(df["Count"])

df2=pd.DataFrame(df2)
df2.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df2["Frequency"]=df2["Count"]/np.sum(df2["Count"])

df3=pd.DataFrame(df3)
df3.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df3["Frequency"]=df3["Count"]/np.sum(df3["Count"])

df4=pd.DataFrame(df4)
df4.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df4["Frequency"]=df4["Count"]/np.sum(df4["Count"])

df5=pd.DataFrame(df5)
df5.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df5["Frequency"]=df5["Count"]/np.sum(df5["Count"])

df6=pd.DataFrame(df6)
df6.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df6["Frequency"]=df6["Count"]/np.sum(df6["Count"])

df7=pd.DataFrame(df7)
df7.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df7["Frequency"]=df7["Count"]/np.sum(df7["Count"])

df8=pd.DataFrame(df8)
df8.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df8["Frequency"]=df8["Count"]/np.sum(df8["Count"])

df9=pd.DataFrame(df9)
df9.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df9["Frequency"]=df9["Count"]/np.sum(df9["Count"])

df10=pd.DataFrame(df10)
df10.rename(columns={0: "Barcode Sequence", 1: "Count"}, inplace=True)
df10["Frequency"]=df10["Count"]/np.sum(df10["Count"])


#save as csv files
df.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_1_S62_true_barcodes.csv")
df2.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_10_S71_true_barcodes.csv")
df3.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_2_S63_true_barcodes.csv")
df4.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_3_S64_true_barcodes.csv")
df5.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_4_S65_true_barcodes.csv")
df6.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_5_S66_true_barcodes.csv")
df7.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_6_S67_true_barcodes.csv")
df8.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_7_S68_true_barcodes.csv")
df9.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_8_S69_true_barcodes.csv")
df10.to_csv("More Barcodes From Fitz/cxt_mg_12dpi_9_S70_true_barcodes.csv")

data=(df, df2, df3, df4, df5, df6, df7, df8, df9, df10)

#compile the data so all sequences go in one column, all read counts go in another column
data = pd.concat(data, ignore_index=True)

data.to_csv("More Barcodes From Fitz/cxt_12dpi_mg_all_true_barcodes_with_duplicates.csv")

#remove duplicates by averaging their frequencies
data2=data.drop(columns=["Count"], inplace=True)
data2=data.groupby(["Barcode Sequence"]).mean()
data2.sort_values(by=["Frequency"], ascending=False, inplace=True)
data2.to_csv("More Barcodes From Fitz/cxt_12dpi_mg_all_true_barcodes_without_duplicates.csv")

print(data)
print(data2)
```
